# Import Libraries

```
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style for plots
sns.set_style("whitegrid")

# Set display options for better readability
pd.set_option("display.max_colwidth", None)
```

# Load Dataset Files

```
# Load file containing dataset features
features = pd.read_csv(
    "Datasets/UNSW-NB15 - Features.csv",
    encoding="latin-1"
)

# Preview DataFrame
features.head()
```

```
# Load all 4 data files
df1 = pd.read_csv(
    "Datasets/UNSW-NB15 - 1.csv",
    header=None,
    low_memory=False
)
df2 = pd.read_csv(
    "Datasets/UNSW-NB15 - 2.csv",
    header=None,
    low_memory=False
)
df3 = pd.read_csv(
    "Datasets/UNSW-NB15 - 3.csv",
    header=None,
    low_memory=False
)
df4 = pd.read_csv(
    "Datasets/UNSW-NB15 - 4.csv",
    header=None,
    low_memory=False
)

# Preview first data file
df1.head()
```

# Combine Dataset Files

```
# Combine all dataframes using concatenation
df = pd.concat(
    [df1, df2, df3, df4],
    ignore_index=True
)

# Preview combined DataFrame
df.head()
```

# Change Column Headers

```
# Get column headers for combined DataFrame from 'features'
column_names = features["Name"].tolist()
print(f"Column names: {column_names}")

# Change column headers
df.columns = column_names

# Preview complete combined DataFrame
df.head()
```
